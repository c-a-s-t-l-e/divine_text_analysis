### Fetch all the url's for the speeches
Reference: https://www.churchofjesuschrist.org/study/general-conference/speakers/russell-m-nelson?lang=eng

```python
import requests
from bs4 import BeautifulSoup

def get_urls():
    url = 'https://www.churchofjesuschrist.org/study/general-conference/speakers/russell-m-nelson?lang=eng'
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')

    urls = []
    for link in soup.find_all('a'):
        url = link.get('href')
        if url and 'nelson' in url:  # check if 'nelson' is in the url
            urls.append(url)

    return urls

if __name__ == "__main__":
  for url in get_urls():
    print("https://www.churchofjesuschrist.org" + url)
```

```python
import requests
from bs4 import BeautifulSoup
import re

def get_urls():
    url = 'https://www.churchofjesuschrist.org/study/general-conference/speakers/russell-m-nelson?lang=eng'
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')

    urls = []
    for link in soup.find_all('a'):
        url = link.get('href')
        if url and 'general-conference' in url and re.search(r'/(04|10|nelson)/', url):
            # check if 'general-conference' is in the url and either '04' or '10' appear
            urls.append("https://www.churchofjesuschrist.org" + url)

    return urls

# if __name__ == "__main__":
#  for url in get_urls():
#    print(url)
```

```python
def get_speech_text(url):
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')

    text = []
    for paragraph in soup.find_all('p'):
        paragraph_text = paragraph.get_text().strip()  # Remove leading/trailing whitespace
        if len(paragraph_text) > 35 and any(c.isalpha() for c in paragraph_text):  # Check length and if contains any alphabetic characters
            text.append(paragraph_text)

    return text
```

```python
urls = get_urls()

speech_paragraphs = []
for url in urls:
  speech_paragraphs.append(url)
  speech_paragraphs.append(get_speech_text(url))
```

```python
import pandas as pd

df = pd.DataFrame(speech_paragraphs)
```

```python
def restructure_dataframe(df):
    column1_rows = df.iloc[::2, 0].reset_index(drop=True)
    column2_rows = df.iloc[1::2, 0].reset_index(drop=True)

    new_data = {'Column1': column1_rows, 'Column2': column2_rows}
    new_df = pd.DataFrame(new_data)
    return new_df

r_df = restructure_dataframe(df)
```

```python
r_df.columns = 'URL', 'Paragraphs'
```

```python
r_df['Speech Number'] = r_df.index
```

```python
r_df.head(2)
```

```python
# "Explode" the 'Paragraphs' column
r_df = r_df.explode('Paragraphs')
```

### Approach No. 1: Phrase Matching

```python
keywords = ['invite', 'encourage', 'ask', 'urge', 'request', 'call on', 'appeal to', 'plea']
pattern = '|'.join(keywords)

invitations = r_df[r_df['Paragraphs'].str.contains(pattern, case=False, na=False)]
```

### Approach No. 2: Build RAG system to find all of President Nelson's invitations

```python
!pip3 install langchain
```

```python
from langchain_community.llms import Ollama
llm = Ollama(model="tinyllama")
```

```python
llm.invoke("how can langsmith help with testing?")
```

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader(urls)
docs = loader.load()
docs
```

```python
from langchain_community.embeddings import OllamaEmbeddings

embeddings = OllamaEmbeddings(model='tinyllama')
```

```python
!pip3 install faiss-cpu
```

```python
from langchain_community.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter


text_splitter = RecursiveCharacterTextSplitter()
documents = text_splitter.split_documents(docs)
vector = FAISS.from_documents(documents, embeddings)
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template("""Answer the following question based only on the provided context:

<context>
{context}
</context>

Question: {input}""")

document_chain = create_stuff_documents_chain(llm, prompt)
```

```python
from langchain.chains import create_retrieval_chain

retriever = vector.as_retriever()
retrieval_chain = create_retrieval_chain(retriever, document_chain)
```

```python
response = retrieval_chain.invoke({"input": "List the most common inviations that President Nelson gives"})
print(response["answer"])

# LangSmith offers several features that can help with testing:...
```
